7) Vamos fazer um programa para verificar quem é o assassino de um crime. Para descobrir o assassino, a polícia faz um pequeno questionário com 5 perguntas onde a resposta só pode ser sim ou não:  


    a. Mora perto da vítima?

    b. Já trabalhou com a vítima?

    c. Telefonou para a vítima?

    d. Esteve no local do crime?

    e. Devia para a vítima?

Cada resposta sim dá um ponto para o suspeito. A polícia considera que os suspeitos com 5 pontos são os assassinos, com 4 a 3 pontos
são cúmplices e 2 pontos são apenas suspeitos, necessitando outras investigações. Valores iguais ou abaixo de 1 são liberados.

```python
# Inicializa a contagem de pontos do suspeito
pontos = 0

# Pergunta ao suspeito as 5 questões
resposta = input("Mora perto da vítima? (sim ou não): ")
if resposta.lower() == "sim":
    pontos += 1

resposta = input("Já trabalhou com a vítima? (sim ou não): ")
if resposta.lower() == "sim":
    pontos += 1

resposta = input("Telefonou para a vítima? (sim ou não): ")
if resposta.lower() == "sim":
    pontos += 1

resposta = input("Esteve no local do crime? (sim ou não): ")
if resposta.lower() == "sim":
    pontos += 1

resposta = input("Devia para a vítima? (sim ou não): ")
if resposta.lower() == "sim":
    pontos += 1

# Determina o resultado com base nos pontos
if pontos == 5:
    print("Você é o assassino!")
elif pontos >= 3:
    print("Você é um cúmplice.")
elif pontos >= 1:
    print("Você é apenas um suspeito.")
else:
    print("Você está liberado.")
```

8) Um produto vai sofrer aumento de acordo com a Tabela 1 abaixo. Faça um programa que peça para o usuário digitar o valor do produto de acordo
com o preço antigo e escreva uma das mensagens da Tabela 2, de acordo com o preço reajustado:

Tabela 1

| Preço Antigo         | % de aumento |
|----------------------|--------------|
| Até 50 reais         | 5%           |
| Entre 50 e 100 reais | 10%          |
| De 100 a 150 reais   | 13%          |
| Acima de 150 reais   | 15%          |

Tabela 2

| Preço Novo            | Mensagem   |
|-----------------------|------------|
| Até 80 reais          | Barato     |
| Entre 80 e 115 reais  | Razoável   |
| Entre 115 e 150 reais | Normal     |
| Entre 150 e 170 reais | Caro       |
| Acima de 170 reais    | Muito Caro |

```python
# Solicita ao usuário o valor do produto com base no preço antigo
preco_antigo = float(input("Digite o valor do produto (em reais) com base no preço antigo: "))

# Determina o percentual de aumento com base na tabela 1
if preco_antigo <= 50:
    percentual_aumento = 5
elif preco_antigo <= 100:
    percentual_aumento = 10
elif preco_antigo <= 150:
    percentual_aumento = 13
else:
    percentual_aumento = 15

# Calcula o preço novo com base no percentual de aumento
preco_novo = preco_antigo * (1 + percentual_aumento / 100)

# Determina a mensagem com base na tabela 2
if preco_novo <= 80:
    mensagem = "Barato"
elif preco_novo <= 115:
    mensagem = "Razoável"
elif preco_novo <= 150:
    mensagem = "Normal"
elif preco_novo <= 170:
    mensagem = "Caro"
else:
    mensagem = "Muito Caro"

# Imprime a mensagem correspondente
print(f"O preço reajustado é de R${preco_novo:.2f}, considerado {mensagem}.")
```

2) Faça um programa que mostre o fatorial de um número digitado.  
_Exemplo_  
número digitado: 5  
resultado esperado: 120

```python
# Solicita ao usuário que digite um número inteiro
n = int(input("Digite um número inteiro para calcular o fatorial: "))

# Verifica se o número é não negativo
if n < 0:
    print("O fatorial não está definido para números negativos.")
else:
    # Inicializa o resultado do fatorial como 1
    resultado = 1

    # Calcula o fatorial de n usando um loop while
    contador = 1
    while contador <= n:
        resultado *= contador
        contador += 1

    # Imprime o resultado
    print(f"O fatorial de {n} é {resultado}")
```

4) Faça um programa em que o usuário digite números quaisquer e encerrará no momento em que o valor 0 seja digitado. Ao final diga qual foi o maior número digitado.

```python
# Inicializa a variável para armazenar o maior número
maior_numero = None

# Solicita ao usuário que digite números até que 0 seja digitado
while True:
    numero = int(input("Digite um número (ou 0 para encerrar): "))

    # Verifica se o número digitado é 0 para encerrar o loop
    if numero == 0:
        break

    # Atualiza o maior número, se necessário
    if maior_numero is None or numero > maior_numero:
        maior_numero = numero

# Verifica se foram digitados números antes de mostrar o resultado
if maior_numero is not None:
    print(f"O maior número digitado foi: {maior_numero}")
else:
    print("Nenhum número foi digitado.")
```

5) Faça um script que peça para o usuário digitar a idade, o salário e o sexo de uma pessoa até que as entradas digitadas sejam válidas.  
  
a. Idade: entre 0 e 150  
b. Salário: maior que 0  
c. Gênero: M, F ou Outro

6) Faça uma calculadora. O usuário deve inserir qual a operação matemática ele deseja realizar e logo em seguida os dois números. O programa deve finalizar apenas quando o usuário digitar a opção "sair" no momento de escolha da operação matemática.

```python
while True:
    # Solicita ao usuário a operação matemática desejada ou se ele deseja sair
    operacao = input("Digite a operação matemática (+, -, *, /) ou 'sair' para encerrar: ")

    # Verifica se o usuário deseja sair
    if operacao.lower() == 'sair':
        print("Calculadora encerrada. Até a próxima!")
        break

    # Verifica se a operação é válida ou se ele não selecinou nada fora dos operadores matematicos
    if operacao not in ('+', '-', '*', '/'):
        print("Operação inválida. Por favor, escolha uma das operações matemáticas (+, -, *, /) ou 'sair'.")
        continue

    # Solicita os dois números para a operação matematica que vamos fazer
    try:
        num1 = float(input("Digite o primeiro número: "))
        num2 = float(input("Digite o segundo número: "))
    except ValueError:
        print("Por favor, insira números válidos.")
        continue

    # Realiza a operação selecionada como +, -, * e etc
    if operacao == '+':
        resultado = num1 + num2
    elif operacao == '-':
        resultado = num1 - num2
    elif operacao == '*':
        resultado = num1 * num2
    elif operacao == '/':
        if num2 == 0:
            print("Não é possível dividir por zero.")
            continue
        resultado = num1 / num2

    # Exibe o resultado da operação para os outros
    print(f"Resultado: {num1} {operacao} {num2} = {resultado}")
```
